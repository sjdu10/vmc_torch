```python
import quimb.tensor as qtn
import numpy as np
np.random.seed(0)

# random PEPS and random configuration
rpeps = qtn.PEPS.rand(4,4,bond_dim=4,phys_dim=2, seed=0)
random_config = np.random.randint(2, size=(16,))
amp = rpeps.isel({i: s for i, s in zip(rpeps.site_inds, random_config)})

# compute environments with equalize_norms=1.0, which did not record the exponents
env_x = amp.compute_x_environments(
    max_bond=16, cutoff=0.0,
    equalize_norms=1.0,
    canonize=True,
    mode='mps',
)
print('bMPS exponents (BUGGY: equalize_norms=1.0):')
for k, v in env_x.items():
    print(k, f'bMPS exponent: {v.exponent}')
print('Contraction results from exact and bMPS:')
print(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())
print(f'Is two contraction equal? {np.allclose(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())}')

print()
"""
# Correct exponents
('xmax', 3) bMPS exponent: 0.0
('xmax', 2) bMPS exponent: 0.0
('xmax', 1) bMPS exponent: 5.256317733784051
('xmax', 0) bMPS exponent: 7.372078963957738
('xmin', 0) bMPS exponent: 0.0
('xmin', 1) bMPS exponent: 0.0
('xmin', 2) bMPS exponent: 5.361989120615633
('xmin', 3) bMPS exponent: 7.309535124102084
"""
exponent_dict = {('xmax', 3): 0.0, ('xmax', 2): 0.0, ('xmax', 1): np.float64(5.256317733784051), ('xmax', 0): np.float64(7.372078963957738), ('xmin', 0): 0.0, ('xmin', 1): 0.0, ('xmin', 2): np.float64(5.361989120615633), ('xmin', 3): np.float64(7.309535124102084)}
for k, v in env_x.items():
    v.exponent = exponent_dict[k]
print('bMPS exponents (manually setting the exponents):')
for k, v in env_x.items():
    print(k, f'bMPS exponent: {v.exponent}')
    
print('Contraction results from exact and bMPS:')
print(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())
print(f'Is two contraction equal? {np.allclose(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())}')

print()

env_x = amp.compute_x_environments(
    max_bond=16, cutoff=0.0,
    # equalize_norms=1.0,
    canonize=True,
    mode='mps',
)
print('bMPS exponents (equalize_norms=False):')
for k, v in env_x.items():
    print(k, f'bMPS exponent: {v.exponent}')
print('Contraction results from exact and bMPS:')
print(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())
print(f'Is two contraction equal? {np.allclose(amp.contract(), (env_x[('xmax',2)]|env_x[('xmin',3)]).contract())}')

```
